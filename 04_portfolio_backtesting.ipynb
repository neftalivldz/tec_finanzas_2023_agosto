```python
fund, err = ek.get_data(['Portfolio(RETO2023_B)'],['TR.PortfolioConstituentName', 'TR.PortfolioWeight'])
fund
```

```python
instruments = fund['Instrument'].tolist()
instruments
```

```python
weights = fund['Portfolio Weight'].tolist()
weights
```

```python
# Python3 code to demonstrate working of
# Last weekday of every month in year
# Using loop + max() + calendar.monthcalendar
import calendar
 
# initializing year
year = 2022
 
# printing Year
print("The original year : " + str(year))
 
# initializing weekday
weekdy = 5
 
# iterating for all months
res = []
for month in range(3, 13):
     
    # max gets last friday of each month of 1997
    res.append(str(year) + "-" + str(month)  + "-" + str(max(week[weekdy]
                       for week in calendar.monthcalendar(year, month)))) 
 
# printing
print("Last weekdays of year : " + str(res))
```

```python
# initializing year
year = 2023
 
# printing Year
print("The original year : " + str(year))
 
# initializing weekday
weekdy = 5
 
# iterating for all months
for month in range(1, 4):
     
    # max gets last friday of each month of 1997
    res.append(str(year) + "-" + str(month)  + "-" + str(max(week[weekdy]
                       for week in calendar.monthcalendar(year, month)))) 
# printing
print("Last weekdays of year : " + str(res))
```

```python
dates = res
dates
```

```python
fund_df, err = ek.get_data(
    instruments = instruments,
    fields = ['TR.ClosePrice(SDate='+ dates[0][0:10]+',EDate='+dates[-1][0:10]+',Frq=M).date',
             'TR.ClosePrice(SDate='+ dates[0][0:10]+',EDate='+dates[-1][0:10]+',Frq=M)']
)
display(fund_df)
```

```python
#Portfolio Asset Allocation. 
fig = go.Figure(data=[go.Pie(labels=instruments, values=weights)])
fig.update_layout(title="Portfolio Asset Allocation")
fig.show()
```

```python
first = fund_df.groupby('Instrument').first()
first
```

```python
fund2 = pd.merge(fund, first, how = "inner", on = 'Instrument')
fund2["Value"] = 1000000
fund2
```

```python
fund_df2 = pd.merge(fund_df, fund2, how = "inner", on = 'Instrument')
fund_df2
```

```python
fund_df2['Valueperinst'] = fund_df2['Value']*fund_df2['Portfolio Weight']*fund_df2['Close Price_x']/fund_df2['Close Price_y']
fund_df2['Growth'] = fund_df2['Close Price_x']/fund_df2['Close Price_y']
fund_df2
```

```python
portfolio = fund_df2.groupby("Date_x")['Valueperinst'].agg(['sum']).reset_index().rename(columns={"Date_x": "Date", "sum": "PortfolioValue"})
portfolio
```

# Get info for a benchmark

```python
df, err=ek.get_data(['IVV'],['TR.NETASSETVAL.date','TR.NETASSETVAL'], {'SDate':'-1Y', 'Edate':'0D'})
df
```

```python
portfolio2 = pd.merge(portfolio, df, how = "left", on = 'Date')
portfolio2
```

```python
portfolio2['Benchmark Return'] = 1000000*(portfolio2['Net Asset Value']/portfolio2.iloc[0]['Net Asset Value'])
portfolio2.set_index('Date').drop(columns=['Instrument', 'Net Asset Value'])
```
