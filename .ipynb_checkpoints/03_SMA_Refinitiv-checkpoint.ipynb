```python
df = ek.get_data('LP68167376', ['TR.FundRollingPerformance.date',
                           'TR.FundRollingPerformance.value'],
            {'Curn':'SGD','RollTimeFrame':'1Y','Interval':'M'})
```

```python
start_date = '2020-01-01'  # Specify the start date
end_date = '2023-09-01'    # Specify the end date

df = ek.get_timeseries(
    'AAPL.O',                # Ticker symbol for Microsoft
    start_date=start_date,
    end_date=end_date,
    interval='daily'       # Fetch monthly prices
)
```

```python
df
```

```python
df['SMA_30'] = ta.SMA(df.CLOSE, timeperiod=30)
```

```python
df
```

```python
for t in [5, 21, 63]:
    df[f'SMA_{t}'] = ta.SMA(df.CLOSE,
                               timeperiod=t)
```

```python
df
```

```python
for t in [5, 21, 63]:
    df[f'DEMA_{t}'] = ta.DEMA(df.CLOSE,
                                timeperiod=t)
```

```python
df
```

```python
ax = df[["CLOSE", "SMA_5","SMA_21","SMA_63"]].plot(figsize=(14, 5), rot=0)
sns.despine()
ax.set_xlabel('');
```

```python
df['rezago_21'] = df['CLOSE'].shift(21)
```

```python
df['rendimiento'] = np.log(df['CLOSE']/df['rezago_21'])
```
