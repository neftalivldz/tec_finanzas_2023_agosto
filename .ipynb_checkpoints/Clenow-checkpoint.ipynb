```python
from datetime import datetime, timedelta
end = datetime.today()
start = end - timedelta(days=(365.2*4))
start_1y = end - timedelta(days=(365))
print(start, end)
```

```python
df = ek.get_timeseries(const_limp[0], start_date = start, end_date = end, fields='CLOSE', interval="monthly")
df[const_limp[0]] = np.log(df['CLOSE'].div(df['CLOSE'].shift(1)))
monthly_returns = df.drop(['CLOSE'], axis=1)
monthly_returns 
```

```python
meses = round(monthly_returns.count()[0])
meses
```

```python
meses = round(monthly_returns.count()[0])
meses
for i in const_limp[1:]:
    w = ek.get_timeseries(i, start_date = start, end_date = end, fields='CLOSE', interval="monthly")
    w[i] = np.log(w['CLOSE'].div(w['CLOSE'].shift(1)))
    w1 = w.drop(['CLOSE'], axis=1)
    p = round(w1.count()[0])
    print(i, p)
    if p == meses:    
        monthly_returns = monthly_returns.join(w1, lsuffix="_left", rsuffix="_right")
```

```python
ICLN, err = ek.get_data('ICLN.O',['TR.ETPConstituentRIC','TR.ETPConstituentWeightPercent'])
```

```python
ICLN_Constituents = ICLN['Constituent RIC'].dropna().tolist()
const_limp = ICLN_Constituents
```

```python
from datetime import datetime, timedelta
end = datetime.today()
start = end - timedelta(days=(365.2*5))
start_1y = end - timedelta(days=(365))
print(start, end)
```

```python
df = ek.get_timeseries(const_limp[0], start_date = start, end_date = end, fields='CLOSE', interval="monthly")
monthly_prices = df.rename(columns={'CLOSE':const_limp[0]})
monthly_prices 
```

```python
meses = round(monthly_prices.count()[0])
for i in const_limp[1:]:
    w = ek.get_timeseries(i, start_date = start, end_date = end, fields='CLOSE', interval="monthly")
    w1 = w.rename(columns={'CLOSE':i})
    p = round(w1.count()[0])
    print(i, p)
    if p == meses:    
        monthly_prices = monthly_prices.join(w1, lsuffix="_left", rsuffix="_right")
```

```python
monthly_prices.to_csv('precios_ejemplo.csv')
```
